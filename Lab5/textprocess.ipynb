```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import collections
import csv
import string
from collections import Counter
from wordcloud import WordCloud

text="";

with open("chi.txt", "r", encoding="utf-8") as file:
   for line in file:      
       if not re.search("    ", line):
        line = line.lower()
        line = line.replace("[i]", "")
        line = line.replace("[v]", "")
        line = line.replace("[s]", "")
        line = line.replace("(j)", "")
        line = line.replace("(c)", "")
        line = line.replace("(t)", "")
        line = line.replace("using", "")
        line = line.replace("based", "")
        line = line.replace("(best paper award)", "")
        line = line.replace("(best paper honorable mention)", "")
        #print(line)
        text=text+line;
       
    
#print(text)
wordcloud = WordCloud(width=1200, height=800).generate(text)

# Display the generated image:

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
